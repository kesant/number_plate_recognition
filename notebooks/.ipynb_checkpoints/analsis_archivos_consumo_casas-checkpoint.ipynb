```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np
import os
import re
```

```python
# Especifica el directorio donde están los archivos CSV
carpeta ="../data/raw/"

# Obtén la lista de todos los archivos en la carpeta
archivos = os.listdir(carpeta)
archivos
```

```python
# Define the columns to keep
columns_to_keep = [
    'Date/Time',
    'Electricity:Facility [kW](Hourly)',
    'Fans:Electricity [kW](Hourly)',
    'Cooling:Electricity [kW](Hourly)',
    'Heating:Electricity [kW](Hourly)',
    'InteriorLights:Electricity [kW](Hourly)',
    'InteriorEquipment:Electricity [kW](Hourly)'
]
# Load and concatenate data from all files
dfs = []
for file in archivos:
    df = pd.read_csv(f"../data/raw/{file}", usecols=columns_to_keep, parse_dates=['Date/Time'],index_col="Date/Time")
    # Agrega un año por defecto (por ejemplo, 2023) a las fechas
    df.index = '2004 ' + df.index
    # Especifica el formato esperado de las fechas
    date_format = '%Y %m/%d %H:%M:%S'
    # Asegúrate de que el índice 'Date/Time' esté en el formato datetime
    df.index = pd.to_datetime(df.index, format=date_format, errors='coerce')
    match = re.match(r'^[^_]+', filename)
    name = match.group(0)
    df["type_building"]=name
    dfs.append(df)
# # Combine all dataframes into one
# combined_df = pd.concat(dfs, axis=0, ignore_index=True)
# # Set Date/Time as index
# combined_df.set_index('Date/Time', inplace=True)
```
